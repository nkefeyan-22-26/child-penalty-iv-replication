```python
sas7bdat_FILEPATH = "/content/drive/MyDrive/ECON5200/Midterm Project/m_d_806.sas7bdat"
df = pd.read_sas(sas7bdat_FILEPATH)
df.head()
```

Method 1

```python
# Step 2: Downcast the data types to shrink the size
# SAS files often load every number as a 64-bit float even when it doesn't need to be. This alone can cut your file size dramatically:
def optimize_df(df):
    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = pd.to_numeric(df[col], downcast='float')
    for col in df.select_dtypes(include=['int64']).columns:
        df[col] = pd.to_numeric(df[col], downcast='integer')
    return df

df_optimized = optimize_df(df.copy())

print(f"Before: {df.memory_usage(deep=True).sum() / 1_000_000:.1f} MB")
print(f"After:  {df_optimized.memory_usage(deep=True).sum() / 1_000_000:.1f} MB")
```

```python
print(df.columns.tolist())
```

```python
# Keep only the columns that are relevant to Angrist & Evans replication.
# This is essential because the raw data is far too large to process
# Core variables for Angrist & Evans replication
# Angrist & Evans variables mapped to YOUR actual column names
cols_to_keep = [
    # Instrument & key regressors
    'SEXK',       # sex of first child (boy1st)
    'SEX2ND',     # sex of second child (boy2nd)
    'KIDCOUNT',   # number of kids (used to construct morekids)

    # Mother characteristics
    'AGEM',       # mother's age
    'RACEM',      # mother's race
    'SPANISHM',   # mother hispanic
    'GRADEM',     # mother's education
    'MARITAL',    # marital status

    # Mother's labor supply outcomes
    'WEEKSM',     # weeks worked by mother
    'HOURSM',     # hours worked by mother
    'INCOME1M',   # mother's wage income
    'INCOME2M',   # mother's other income

    # Father/spouse characteristics & labor supply
    'AGED',       # father's age
    'RACED',      # father's race
    'GRADED',     # father's education
    'WEEKSD',     # weeks worked by father
    'HOURSD',     # hours worked by father
    'INCOME1D',   # father's wage income

    # Household
    'FAMINC',     # family income
    'STATE',      # state
]

# Only keep columns that actually exist
cols_existing = [c for c in cols_to_keep if c in df.columns]
print(f"Keeping {len(cols_existing)} of {len(df.columns)} columns")

df_slim = df[cols_existing].copy()
print(f"Slim dataframe memory: {df_slim.memory_usage(deep=True).sum() / 1_000_000:.1f} MB")
```

```python
df_slim.to_csv("/content/census80_slim.csv.gz", index=False, compression='gzip')

import os
size = os.path.getsize("/content/census80_slim.csv.gz")
print(f"Final file size: {size / 1_000_000:.1f} MB")
```

Method 2

```python
# Step 1: Convert low-cardinality columns to category dtype
for col in df.columns:
    n_unique = df[col].nunique()
    if n_unique < 50:  # if a column has fewer than 50 unique values, make it categorical
        df[col] = df[col].astype('category')
        print(f"{col}: {n_unique} unique values → category")

# Step 2: Downcast all remaining numeric columns
for col in df.select_dtypes(include=['float64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='float')
for col in df.select_dtypes(include=['int64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='integer')

print(f"\nOptimized memory usage: {df.memory_usage(deep=True).sum() / 1_000_000:.1f} MB")
```

```python
# Step 3: Save locally with maximum gzip compression
# compression level 9 is the slowest but smallest possible file
df.to_csv("/content/census80_full.csv.gz",
          index=False,
          compression={'method': 'gzip', 'compresslevel': 9})

size = os.path.getsize("/content/census80_full.csv.gz")
print(f"Final file size: {size / 1_000_000:.1f} MB")
```
